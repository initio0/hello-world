```python
import pandas as pd
from datetime import date
import numpy as np


df = pd.read_excel("~/Downloads/mkt_data.xlsx", index_col="Date", parse_dates=True).sort_index(ascending=True)

df.head()
```

```python
df["d_US10Y"] = df["US10Y"].diff()*100
df["d_SPX"] = df["SPX"].pct_change()*100
df["d_CDXIG5Y"] = df["CDXIG5Y"].diff()
df["d_WTI"] = df["WTI"].pct_change()*100
df["d_UMBS30OAS"] = df["UMBS30OAS"].diff()

df.head(10)
```

```python
df.index
```

```python

start_date = pd.to_datetime("2022-01-01")
today = pd.to_datetime(date.today())
  
dff = df[(df.index >= start_date) & (df.index <= today)]

dff.columns
```

```python
idx = np.abs(dff['d_US10Y']).sort_values(ascending=False).index

print(idx[0:10])

dff.reindex(idx).head(10)
```

```python
import pandas as pd

def show_dff(dff, idx=None):
    """
    Displays a DataFrame with an optional index order.

    Args:
        dff (pd.DataFrame): The DataFrame to display.
        idx (list, optional): A list of index values to order the DataFrame by. Defaults to None.

    Returns:
        pd.DataFrame: The displayed DataFrame.
    """
    if idx is not None:
        dff = dff.reindex(idx)
    return dff
```
